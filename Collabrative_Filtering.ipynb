```
df=pd.read_csv("/content/drive/My Drive/Dataset/clean.csv",index_col=0)
df.head(5)
```

```
d=df.T
d.head(5)
```

```
corrMatrix = d.corr(method='pearson')
corrMatrix.head(5)
```

```
corrMatrix[10246]
```

```
def get_similar(user_id):
    similar_score = corrMatrix[user_id]
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score
```

```
user = [(22534)]
similar_scores = pd.DataFrame()
for user_id in user:
    similar_scores = similar_scores.append(get_similar(user_id),ignore_index = True)

similar_scores.head(10)
```

```
similar_scores.sum().sort_values(ascending=False)
```

```
p=pd.DataFrame(similar_scores.sum().sort_values(ascending=False),columns=["similarity"])
p
```

```
disimiler=p[p['similarity'] < 0].index
p.drop(disimiler , inplace=True)
p
```
